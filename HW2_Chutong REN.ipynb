```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation 微分方程ODE
def ode_rhs_1(V, t):
    return -10 - V - t

def ode_rhs_2(V, t):
    return np.cos(t) - 1/2 * V + 20

# Define the axis limits
t0 = -6
t1 = 10
V0 = -20
V1 = 0

# Create a grid of time and voltage values
t = np.linspace(t0, t1, 30) # 起始值，终止值，长度为 30，等间隔序列 时间
V = np.linspace(V0, V1, 30) # 电压
t_grid, V_grid = np.meshgrid(t, V)  #两个一位数组，转换成一个二维网格，每个点由（t, V）组成

# Calculate the slope vectors
delta_t = (t1 - t0) / 30  #为啥除以30
delta_V = (V1 - V0) / 30
dt = 2 * delta_t
dV = ode_rhs_1(V_grid, t_grid) * delta_t 

# Normalize the vectors
length = np.sqrt(dt**2 + dV**2)
dt /= length
dV /= length

# 绘制斜率场
plt.quiver(t, V, dt, dV)

plt.contour(t, V, ode_rhs_1(V_grid, t_grid), levels=[-3], colors='red', linestyles='solid', label='isocline -3')
plt.contour(t, V, ode_rhs_1(V_grid, t_grid), levels=[-1], colors='orange', linestyles='solid', label='isocline -1')
plt.contour(t, V, ode_rhs_1(V_grid, t_grid), levels=[1], colors='purple', linestyles='solid', label='isocline 1')

legend_elements = [
    Line2D([0], [0], color='red', linestyle='solid', label='isocline -3 V/s'),
    Line2D([0], [0], color='orange', linestyle='solid', label='isocline -1 V/s'),
    Line2D([0], [0], color='purple', linestyle='solid', label='isocline 1 V/s')
]
legend = plt.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.0, 1.0), framealpha=1.0)
legend.get_frame().set_facecolor('white')
plt.xlabel('t(s)')
plt.ylabel('V(V)')
plt.yticks(np.arange(-20, 1, 2))
plt.rcParams.update({'font.size': 17}) 
plt.show()
```
